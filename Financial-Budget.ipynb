```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv(r"India_budget_2021.csv")
```

```python
data = data.dropna()
data
```

# Which Ministry has highest Budget Allocation 

```python
plt.figure(figsize=(18,8))
sns.barplot(data=data,x="Department /Ministry",y="Fund allotted(in â‚¹crores)",palette="viridis")
plt.xticks(rotation=90)
plt.title("Graph Between Fund Allocated And Department / Ministry")
plt.show()
```

```python
data=data.iloc[:-1 , :]
data
```

```python
plt.figure(figsize=(18,8))
sns.barplot(data,x='Department /Ministry',y='Fund allotted(in â‚¹crores)',palette="spring")
plt.xticks(rotation=90)
# plt.xlabel(fontsize=10)
plt.title("Graph Between Fund Allocated And Department / Ministry",fontsize=20,weight="bold")
plt.show()
```

# Distribution of Budget For Top 10 Ministries

```python
data = data.sort_values(by='Fund allotted(in â‚¹crores)',ascending=False)
data.head(10)
```

```python
new_data = data.head(10)
df = new_data["Fund allotted(in â‚¹crores)"]
labels=new_data['Department /Ministry']
plt.figure(figsize=(8,7))
plt.pie(df,labels=labels,autopct="%1.1f%%",startangle=45,pctdistance=0.7,shadow=True)
central_circle = plt.Circle((0,0),0.5,color="white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font',size=12)
plt.title("Distribution of Budget in Top 10 Ministries",weight="bold")
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.barplot(data,x='Department /Ministry',y='Fund allotted(in â‚¹crores)',palette="hot")
plt.xticks(rotation=90)

plt.title("Distribution of Budget in Top 10 Ministries",fontsize=20,weight="bold")
plt.show()
```

# Distriution of Budget For Last 10 Ministries 

```python
data = data.sort_values(by='Fund allotted(in â‚¹crores)',ascending=True)
data.head(10)
```

```python
new_data = data.head(10)
df = new_data["Fund allotted(in â‚¹crores)"]
labels=new_data['Department /Ministry']
plt.figure(figsize=(8,7))
plt.pie(df,labels=labels,autopct="%1.1f%%",startangle=0,pctdistance=0.7,shadow=True)
central_circle = plt.Circle((0,0),0.8,color="white")
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font',size=12)
plt.title("Distribution of Budget in Last 10 Ministries",weight="bold")
plt.show()
```

```python
plt.figure(figsize=(18,8))
sns.barplot(data,x='Department /Ministry',y='Fund allotted(in â‚¹crores)',palette="spring")
plt.xticks(rotation=90)

plt.title("Distribution of Budget in Last 10 Ministries",fontsize=20,weight="bold")
plt.show()
```

# We can Also place it Between last 10 ministries
